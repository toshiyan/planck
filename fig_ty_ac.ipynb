```python
import numpy as np
from matplotlib.pyplot import *
```

```python
Dir = '/global/homes/t/toshiyan/scratch/plk/xcor/aps/'
```

```python
def plotM(Dir,t='smica',y='milca'):
    for M in [1,2,3,4]:
        b, cb, vb = np.loadtxt(Dir+'xl_obs_TT_bhlens_'+t+'_Lmask_a1.0deg_l100-2048_'+y+'_M'+str(M)+'_oL1-2048_b10.dat',unpack=True,usecols=(0,1,4))
        fac = b**2/2./np.pi
        errorbar(b+M*20,cb*fac,yerr=vb*fac,fmt='o',label='M'+str(M))
```

```python
def plotTY(Dir,M=1):
    for i, t in enumerate(['smica','nilc']):
        b, cb, vb = np.loadtxt(Dir+'xl_obs_TT_bhlens_'+t+'_Lmask_a1.0deg_l100-2048_milca_M'+str(M)+'_oL1-2048_b10.dat',unpack=True,usecols=(0,1,4))
        fac = b**2/2./np.pi
        errorbar(b+i*20,cb*fac,yerr=vb*fac,fmt='o',label='CMB map:'+str(t)+', ymap:milca')
    for i, y in enumerate(['nilc','milca']):
        b, cb, vb = np.loadtxt(Dir+'xl_obs_TT_bhlens_smica_Lmask_a1.0deg_l100-2048_'+y+'_M'+str(M)+'_oL1-2048_b10.dat',unpack=True,usecols=(0,1,4))
        fac = b**2/2./np.pi
        errorbar(b+i*20+40,cb*fac,yerr=vb*fac,fmt='o',label='CMB map:smica, y map:'+str(y))
```

```python
xlabel(r'$L$')
ylabel(r'$L^2C^{\tau y}_L/2\pi$')
plotTY(Dir)
#plotM(Dir)
axhline(0,color='k',ls='--')
legend(loc=0)
ylim(-1.2e-8,1.2e-8)
savefig('fig_ac.png')
show()
```
