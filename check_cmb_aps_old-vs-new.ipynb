```python
import numpy as np, curvedsky, prjlib, plottools as pl, binning as bn, healpy as hp
from matplotlib.pyplot import *
```

```python
p = prjlib.init_analysis(wtype='Lmask',ascale=1.0)
```

```python
dold = '/global/homes/t/toshiyan/scratch/PLKold/'
cl = {}
CL = {}
L, CL['r'] = np.loadtxt(dold+'aps/aps_obs_1d_smica_Lmask1.dat',usecols=(0,1),unpack=True)
L, CL['s'], CL['n'], CL['p'], CL['c'] = np.loadtxt(dold+'aps/aps_sim_1d_smica_Lmask1.dat',usecols=(0,1,2,3,4),unpack=True)
l, cl['r'] = np.loadtxt(p.fcmb.ocl,unpack=True,usecols=(0,1))
l, cl['s'], cl['n'], cl['p'], cl['c'] = np.loadtxt(p.fcmb.scl,unpack=True,usecols=(0,1,2,3,4))
```

```python
for s in ['s','n','p','c','r']:
    pl.plot_1dstyle(fsize=[20,4],ymin=1e-11,ymax=5e-9,xmin=2,xmax=2048)
    plot(L,L**2*CL[s],label='old')
    plot(l,l**2*cl[s],label='new')
    legend()
    show()
```

```python
for s in ['s','n','p','c','r']:
    pl.plot_1dstyle(fsize=[20,4],ymin=0.95,ymax=1.1,xmin=2,xmax=2048)
    axhline(1,color='k',ls='--')
    plot(l[1:],CL[s][0:p.lmax]/cl[s][1:],label=s)
    legend()
    show()
```
