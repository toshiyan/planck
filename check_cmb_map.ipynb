```python
import numpy as np, healpy as hp, prjlib, pickle, curvedsky, plottools as pl, tools_cmb
from matplotlib.pyplot import *
%matplotlib inline
```

```python
p = prjlib.init_analysis(wtype='Lmask',fltr='cinv')
q = prjlib.init_analysis(wtype='Lmask')
r = prjlib.init_analysis(wtype='Fullsky')
```

1. Inverse Noise variance map

```python
Nij = pickle.load(open(q.fcmb.fnij,"rb"))
```

```python
hp.mollview(Nij[0,0,:])
```

2. Cinv filtered map

```python
# real data
Tmap = {}
Talm = pickle.load(open(p.fcmb.alms['o']['T'][0],"rb"))
Tmap['cinv'] = curvedsky.utils.hp_alm2map(1024,p.lmax,p.lmax,Talm)
Talm = pickle.load(open(q.fcmb.alms['o']['T'][0],"rb"))
Tmap['none'] = curvedsky.utils.hp_alm2map(1024,q.lmax,q.lmax,Talm)
pl.view_maps(Tmap,min=-5e-5,max=5e-5,M=1.)
```

```python
# sim
Tmap = {}
Talm = pickle.load(open(p.fcmb.alms['o']['T'][1],"rb"))
Tmap['cinv'] = curvedsky.utils.hp_alm2map(1024,p.lmax,p.lmax,Talm)
Talm = pickle.load(open(q.fcmb.alms['o']['T'][1],"rb"))
Tmap['none'] = curvedsky.utils.hp_alm2map(1024,q.lmax,q.lmax,Talm)
Talm = pickle.load(open(r.fcmb.alms['s']['T'][1],"rb"))
Tmap['full'] = curvedsky.utils.hp_alm2map(1024,q.lmax,q.lmax,Talm)
pl.view_maps(Tmap,min=-5e-5,max=5e-5,M=1.)
```

3. Mean value of SMICA

```python
hpmap = tools_cmb.reduc_map(p.dtype,p.fimap['s'][0],nside=1024)
hp.mollview(hpmap)
print(np.mean(hpmap))
```
