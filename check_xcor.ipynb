```python
import numpy as np
import curvedsky
import prjlib
from matplotlib.pyplot import *
import plottools as pl
```

```python
ytype = 'nilc'
p = prjlib.init_analysis(fltr='none')
qtau, qlen, qtbh, qlbh = prjlib.init_quad(p.ids,p.stag,rlmin=100,rlmax=2048)
cy = prjlib.init_compy(p.ids,masktype=0,ytype=ytype)
fxtau, fxtbh = prjlib.init_cross(qtau,qtbh,cy,p.ids,p.stag)
```

```python
b, y0, y1, y2, xl = np.loadtxt(cy.fclyy,unpack=True,usecols=(0,1,2,3,4))
s = b**2/2/np.pi*1e12
pl.plot_1dstyle(xmin=3,xmax=2000,ymin=1e-3,ymax=10,xlog=True,ylog=True,grid=True,ylab=r'$10^{12}\times L^2C^{yy}_L/2\pi$')
plot(b,s*y0,label='F+L')
plot(b,s*xl,label='F/L')
legend()
```

```python
#l, xsl = np.loadtxt(fxtau.mxls,unpack=True,usecols=(0,1))
b, xob = np.loadtxt(fxtau.oxbs,unpack=True,usecols=(0,1))
b, xsb, vsb = np.loadtxt(fxtau.mxbs,unpack=True,usecols=(0,1,4))
```

```python
pl.plot_1dstyle(ymin=-5e-8,ymax=5e-8)
#plot(l,l**2*xsl,label='cross')
plot(b,b**2*xsb,label='cross')
errorbar(b,b**2*xob,yerr=b**2*vsb,label='cross')
legend()
```

```python
b, Xob = np.loadtxt(fxtbh.oxbs,unpack=True,usecols=(0,1))
b, Xsb, Vsb = np.loadtxt(fxtbh.mxbs,unpack=True,usecols=(0,1,4))
pl.plot_1dstyle(ymin=-3e-9,ymax=1e-8)
plot(b,b**2*Xsb/2/np.pi,label='cross')
errorbar(b,b**2*Xob/2/np.pi,yerr=b**2*Vsb/2/np.pi,fmt='o',label='cross')
legend()
```

```python
l, Xsl = np.loadtxt(fxtbh.mxls,unpack=True,usecols=(0,1))
b, Xob = np.loadtxt(fxtbh.oxbs,unpack=True,usecols=(0,1))
b, Xsb = np.loadtxt(fxtbh.mxbs,unpack=True,usecols=(0,1))
```

```python
pl.plot_1dstyle(ymin=-5e-8,ymax=5e-8)
plot(l,l**2*Xsl,label='cross')
plot(b,b**2*Xsb,label='cross')
plot(b,b**2*Xob,label='cross')
legend()
```
