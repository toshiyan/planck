```python
dold = '/global/homes/t/toshiyan/scratch/PLKold/'
L, AL = np.loadtxt(dold+'tderiv/Al1d_smica_Lmask1_l100-2048_oL1-2048.dat',usecols=(0,1),unpack=True)
L, AL0 = np.loadtxt(dold+'old/tderiv/Al1d_smica_Lmask1_l100-2048_oL1-2048.dat',usecols=(0,1),unpack=True)
L, N0 = np.loadtxt(dold+'tderiv/N0_smica_Lmask1_l100-2048.dat',usecols=(0,1),unpack=True)
iN = np.loadtxt(dold+'tderiv/n0_smica_Lmask1_l100-2048.dat',unpack=True)[1]
l, al = np.loadtxt(qtau0.f['TT'].al,unpack=True,usecols=(0,1))
l, n0 = np.loadtxt(qtau0.f['TT'].n0bs,unpack=True,usecols=(0,1))
yscale('log')
plot(L,AL)
plot(L,AL0)
plot(L,N0)
```
